# 1. Importing libraries

```python
import pandas as pd
import pandas.util.testing as tm

import numpy as np

from pylab import rcParams
rcParams['figure.figsize'] = 12, 8

import re

from datetime import timedelta, date

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

from scipy import stats

import seaborn as sns

from dateutil.relativedelta import relativedelta

from textblob import TextBlob
from textblob import Word


# WordCloud - Python linrary for creating image wordclouds
from wordcloud import WordCloud

%matplotlib inline
```

# 2. Importing dataset

```python
col_list = ["date", "tweet"]

df = pd.read_csv("dataset/final.csv", usecols=col_list)
```

```python
# Print dataframe
#df.style.set_properties(subset=['tweet'], **{'width': '800px'}).hide_index()
df.dropna(subset = ["tweet"], inplace=True)
df
```

# 3. Polarity distribution

```python
# Calculating Subjectivity of each

def getTextSubjectivity(txt):
    return TextBlob(txt).sentiment.subjectivity

df['Subjectivity'] = df['tweet'].apply(getTextSubjectivity)
```

```python
# Calculating Polarity of each tweet
def getTextPolarity(txt):
    return TextBlob(txt).sentiment.polarity

df['Polarity'] = df['tweet'].apply(getTextPolarity)
```

```python
# negative, nautral, positive analysis
def getTextAnalysis(a):
    if a < 0:
        return "Negative"
    elif a == 0:
        return "Neutral"
    else:
        return "Positive"

df['Score'] = df['Polarity'].apply(getTextAnalysis)
```

```python

df
```

```python
# calculating the percentage of positive tweets
positive = df[df['Score'] == 'Positive']

print(str(positive.shape[0]/(df.shape[0])*100) + " % of positive tweets")
```

```python
# plotting the tweet polarity

labels = df.groupby('Score').count().index.values

values = df.groupby('Score').size().values

plt.bar(labels, values)
```

```python
for index, row in df.iterrows():
    if row['Score'] == 'Positive':
        plt.scatter(row['Polarity'], row['Subjectivity'], color="green")
    elif row['Score'] == 'Negative':
        plt.scatter(row['Polarity'], row['Subjectivity'], color="red")
    elif row['Score'] == 'Neutral':
        plt.scatter(row['Polarity'], row['Subjectivity'], color="blue")

plt.title('Twitter tweets Analysis')
plt.xlabel('Polarity')
plt.ylabel('Subjectivity')
plt.figure(figsize=(20,10))
# add legend
plt.show()
```

```python
for index, row in df.iterrows():
    if row['Score'] == 'Positive':
        plt.scatter(row['date'], row['Polarity'], color="green")
    elif row['Score'] == 'Negative':
        plt.scatter(row['date'], row['Polarity'], color="red")
    elif row['Score'] == 'Neutral':
        plt.scatter(row['date'], row['Polarity'], color="blue")
#plt.setp(ax.get_xticklabels(), rotation=45)
plt.title('Twitter Sentiment Analysis - Polarity Distribution')
plt.xlabel('Date')
plt.ylabel('Polarity')

plt.figure(figsize=(40,20))

# add legend
plt.show()
```

```python
objective = df[df['Subjectivity'] == 0]

print(str(objective.shape[0]/(df.shape[0])*100) + " % of objective tweets")
```

```python
words = ' '.join([tweet for tweet in df['tweet']])
wordCloud = WordCloud(width=600, height=400).generate(words)
plt.figure(figsize=(20,10))
plt.imshow(wordCloud)
plt.show()
```
