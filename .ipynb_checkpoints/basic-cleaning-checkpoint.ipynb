# 1. Importing libraries

```python
import pandas as pd
import numpy as np
import re

```

# 2. Importing dataset

```python
col_list = ["date", "tweet"]

df = pd.read_csv("dataset/twitter-data-1.csv", usecols=col_list)
```

```python
# Print dataframe
df.style.set_properties(subset=['tweet'], **{'width': '800px'}).hide_index()
```

# 3. Data Cleaning - Basic 

```python

# 3.1 Drop duplicate rows
df.drop_duplicates(subset=None, inplace=True)

# 3.2 Remove twitter pictures links
#df.tweet.remove_pic_links(df)
#df['tweet'] = df['tweet'].str.replace('pic.twitter.com','')
#df.tweet.replace(df.tweet.str.startswith('pic.twitter', na=False): 'pic.twitter','')
df['tweet'] = df.tweet.str.replace(r'(^.*pic.twitter.*$)', '')

# 3.3 Remove 'https' links
df.tweet.apply(lambda x: x.replace('(https?:\/\/(?:www\.|(?!www))[a-zA-Z0-9][a-zA-Z0-9-]+[a-zA-Z0-9]\.[^\s]{2,}|www\.[a-zA-Z0-9][a-zA-Z0-9-]+[a-zA-Z0-9]\.[^\s]{2,}|https?:\/\/(?:www\.|(?!www))[a-zA-Z0-9]+\.[^\s]{2,}|www\.[a-zA-Z0-9]+\.[^\s]{2,})',''))
df['tweet'] = df.tweet.str.replace(r'(^.*http.*$)', '')

# 3.4 Case conversion
#df['tweet'] = df['tweet'].apply(lambda x: " ".join(x.lower() for x in x.split()))

# 3.5 Remove punctuations 
#df['tweet'] = df['tweet'].str.replace('[^\w\s]','')

# 3.6 Remove hashtags

df['tweet'] = df.tweet.str.replace(r'(^.*#.*$)', '')
#df['tweet'] = df['tweet'].str.replace('#','')

# 3.7 Remove '\n', which is auto-inserted when rows are combined
df['tweet'] = df['tweet'].str.replace('\n','')

```

```python
# print dataframe
df\
.style.set_properties(subset=['tweet'], **{'width': '800px'}).hide_index()
```

```python
# Write to a new csv file
df.groupby('date')['tweet'].apply(' '.join).reset_index().to_csv("dataset/basic-cleaning.csv", mode='w', index=None)
```
